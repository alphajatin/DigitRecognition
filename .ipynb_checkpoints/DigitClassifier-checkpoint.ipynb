```python
import pandas as pd
import matplotlib.pyplot as plt, matplotlib.image as mpimg
from sklearn.model_selection import train_test_split
from sklearn import svm
%matplotlib inline
```

```python
labeled_images = pd.read_csv('../DigitRecognition/input/train.csv')
images = labeled_images.iloc[0: ,1:]
labels = labeled_images.iloc[0: ,:1]
train_images, test_images, train_labels, test_labels = train_test_split(images, labels, train_size=0.8, test_size=0.2, random_state=0)
```

```python
i = 1
img = train_images.iloc[i].values
img = img.reshape((28,28))
plt.imshow(img, cmap="gray")
plt.title(train_labels.iloc[i,0])
```

```python
plt.hist(train_images.iloc[i])
```

```python
clf = svm.SVC()
clf.fit(train_images, train_labels.values.ravel())
clf.score(test_images, test_labels)
```

```python
test_images[test_images>0]=1
train_images[train_images>0]=1

img=train_images.iloc[i].values.reshape((28,28))
plt.imshow(img, cmap="binary")
plt.title(train_labels.iloc[i])
```

```python
plt.hist(train_images.iloc[i])
```

```python
clf = svm.SVC()
clf.fit(train_images, train_labels.values.ravel())
clf.score(test_images, test_labels)
```

```python
test_data = pd.read_csv("../DigitRecognition/input/test.csv")
test_data[test_data>0]=1
results=clf.predict(test_data[0:])
```

```python
results
```

```python
df = pd.DataFrame(results)
df.index.name="ImageId"
df.index+=1
df.columns=["Label"]
df.to_csv("results.csv", header=True)
```
